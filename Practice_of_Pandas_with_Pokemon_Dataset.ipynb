## Making Changes to the data

```python
# Adding new columns

pokemon['Total_Power'] = pokemon['HP'] + pokemon['Attack'] + pokemon['Defense'] + pokemon['Sp. Atk'] + pokemon['Sp. Def'] + pokemon['Speed']
pokemon.head(10)


df_copy = pokemon.copy()
```

```python
#dropping columns

df1 = pokemon.drop(columns = ['HP', 'Attack', 'Defense', 'Sp. Atk' , 'Sp. Def' , 'Speed'], axis =1)
df1
```

```python
# another way of adding a new column

pokemon['Total'] = pokemon.iloc[: , 4:10].sum(axis  = 1)
pokemon.head(10)
```

### Rearranging the columns

```python
cols = list(pokemon.columns)

pokemon = pokemon[ cols[0:4] + [cols[-1]] + cols[4:12]]
pokemon.head(10)
```

###  Filtering Data

```python
new_df = pokemon.loc[(pokemon['Type 1'] == 'Grass') & (pokemon['Type 2'] == 'Poison') & (pokemon['HP'] > 70)]
# new_df.to_csv('filtered_df.csv', index = False)

# new_df = new_df.reset_index()
# new_df

new_df.reset_index(drop = True , inplace = True)
new_df
```

```python
pokemon.loc[pokemon['Name'].str.contains('Mega')  &  (pokemon['Sp. Atk'] > 130) &  (pokemon['Type 1'] == 'Fire')]
```

```python
# using regex for a bit complex filtering

import re

pokemon.loc[pokemon['Type 1'].str.contains('fire|grass', flags = re.I, regex = True)]
```

```python
pokemon[pokemon['Name'].str.contains('^pi[a-z]*', flags = re.I , regex = True)]
```

### Conditional Changes

```python
pokemon.loc[pokemon['Type 1'] == 'Fire', 'Type 1'] = 'Flame'
pokemon
```

```python
pokemon.loc[pokemon['Type 1'] == 'Fire', 'Legendary'] = True
pokemon
```

```python
pokemon.loc[pokemon['Total'] > 500 , ['Generation' , 'Legendary']] = 'TEST VALUE'

pokemon
```

```python
pokemon.loc[pokemon['Total'] > 500, ['Generation' , 'Legendary']] = ['Test 1', 'Test 2']
pokemon
```

## Aggregate Statistics (Groupby)

```python
pokemon.groupby(['Type 1']).mean().sort_values( by = 'Attack', ascending = False)
```

```python
pokemon.groupby(['Type 1']).sum()
```

```python
pokemon['count'] = 1

pokemon.groupby(['Type 1', 'Type 2']).count()['count']
```

## Working with large amount of data

```python
new_df = pd.DataFrame(columns = pokemon.columns)


for df in pd.read_csv('/content/pokemon_data.csv', chunksize = 20):
  # print("CHUNK SIZE")
  # print(df)
  results = df.groupby(['Type 1']).count()

  new_df = pd.concat([new_df, results ])
```
